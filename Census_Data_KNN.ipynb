
***1. Replace ? to null for the whole dataset***

```python
df_adult.replace('?', np.nan, inplace=True)
df_adult['income'] = df_adult['income'].str.replace('.', '')
```

***2. Trim leading spaces***

```python
df_adult = df_adult.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df_adult.head(3)
```

```python
df_adult['education'] = df_adult['education'].apply(lambda x: 'School' if x == '11th' or x == '7th-8th' or x == '10th'
                                              or x == '5th-6th' or x == '9th' or x == '12th' or x == '1st-4th'
                                              or x == 'Preschool' else x)
df_adult['education'] = df_adult['education'].apply(lambda x: 'Associate' if x == 'Assoc-acdm' or x == 'Assoc-voc' else x)
education_map = {'School':1,
             'HS-grad':2,
             'Some-college':3,
             'Bachelors':4,
             'Prof-school':5,
             'Associate':6,
             'Masters':7,
             'Doctorate':8}
df_adult['education'] = df_adult['education'].map(education_map)
```

```python
missingvalues_percentage = df_adult.isna().sum().sort_values(ascending= False)/len(df_adult)
missingvalues_percentage
```

```python
from sklearn.impute import KNNImputer
```

```python
from sklearn.model_selection import train_test_split
import category_encoders as ce
```

```python
X = df_adult.drop(['income'], axis=1)

y = df_adult['income']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

```python
encoder = ce.OrdinalEncoder(cols=['workclass', 'education', 'marital-status', 'relationship', 'race', 'sex','native-country','occupation'])


X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=4).fit(X_train,y_train)
```

```python
y_hat = neigh.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_hat)
```

```python
Ks=30
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    y_hat = neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test,y_hat)
    std_acc[n-1] = np.std(y_hat==y_test)/np.sqrt(y_hat.shape[0])
    
mean_acc
```

```python
with open("MeanAccKNN.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    
    writer.writerow(["Serial Number", "Value"])
    
    for i, value in enumerate(mean_acc, start=1):
        writer.writerow([i, value])
```

```python
plt.plot(range(1,Ks),mean_acc,'g',marker='o')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Nabors (K)')
plt.tight_layout()
plt.show()
```

```python
neigh = KNeighborsClassifier(n_neighbors=12).fit(X_train,y_train)
y_hat = neigh.predict(X_test)
metrics.accuracy_score(y_test,y_hat)
```

```python
from sklearn.metrics import confusion_matrix, confusion_matrix
confusion_matrix(y_test, y_hat)
```

```python
y_pred = neigh.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Prediction')
plt.show()
```
